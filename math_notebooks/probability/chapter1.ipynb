## 1. ä¹˜æ³•åŸç† (Multiplication Rule)

### ğŸ“– ç†è«–

**ä¹˜æ³•åŸç† (Multiplication Principle)**ï¼š
å¦‚æœä¸€å€‹å¯¦é©—æœ‰ $n$ å€‹æ­¥é©Ÿï¼Œç¬¬ 1 æ­¥æœ‰ $n_1$ ç¨®é¸æ“‡ï¼Œç¬¬ 2 æ­¥æœ‰ $n_2$ ç¨®é¸æ“‡ï¼Œ...ï¼Œç¬¬ $n$ æ­¥æœ‰ $n_n$ ç¨®é¸æ“‡ï¼Œå‰‡ç¸½å…±æœ‰ï¼š

$$\text{ç¸½æ•¸} = n_1 \times n_2 \times \cdots \times n_n$$

### ğŸ² ä¾‹å­ï¼šå†°æ·‡æ·‹åº—

å‡è¨­ä¸€å®¶å†°æ·‡æ·‹åº—æœ‰ï¼š
- 3 ç¨®å£å‘³ (flavors): å·§å…‹åŠ›ã€é¦™è‰ã€è‰è“
- 2 ç¨®å®¹å™¨ (containers): ç”œç­’ã€æ¯å­
- 2 ç¨®é…æ–™ (toppings): æœ‰ã€ç„¡

**å•é¡Œ**ï¼šç¸½å…±æœ‰å¹¾ç¨®çµ„åˆï¼Ÿ

**ç­”æ¡ˆ**ï¼š$3 \times 2 \times 2 = 12$ ç¨®

```python
import itertools
import math
from typing import List

# ============================================
# 1. ä¹˜æ³•åŸç†ï¼šå†°æ·‡æ·‹åº—ä¾‹å­
# ============================================

# å®šç¾©é¸é …
flavors = ['Chocolate', 'Vanilla', 'Strawberry']
containers = ['Cone', 'Cup']
toppings = ['With Topping', 'No Topping']

# æ–¹æ³•ä¸€ï¼šä½¿ç”¨ itertools.product åˆ—èˆ‰æ‰€æœ‰çµ„åˆ
all_combinations = list(itertools.product(flavors, containers, toppings))

print("=" * 60)
print("ğŸ¦ å†°æ·‡æ·‹çµ„åˆ (Ice Cream Combinations)")
print("=" * 60)
print(f"å£å‘³ (Flavors): {len(flavors)} ç¨®")
print(f"å®¹å™¨ (Containers): {len(containers)} ç¨®")
print(f"é…æ–™ (Toppings): {len(toppings)} ç¨®")
print(f"\næ ¹æ“šä¹˜æ³•åŸç†ï¼š{len(flavors)} Ã— {len(containers)} Ã— {len(toppings)} = {len(flavors) * len(containers) * len(toppings)} ç¨®")
print(f"å¯¦éš›åˆ—èˆ‰çµæœï¼š{len(all_combinations)} ç¨®")
print("\næ‰€æœ‰çµ„åˆï¼š")
print("=" * 60)

for i, combo in enumerate(all_combinations, 1):
    flavor, container, topping = combo
    print(f"{i:2d}. {flavor:12} | {container:4} | {topping}")
    
print("=" * 60)
```

## 2. æ’åˆ— (Permutations)

### ğŸ“– ç†è«–

**æ’åˆ— (Permutation)**ï¼šå¾ $n$ å€‹ä¸åŒçš„ç‰©ä»¶ä¸­ï¼Œé¸å‡º $k$ å€‹é€²è¡Œæ’åˆ—ï¼ˆé †åºé‡è¦ï¼‰ã€‚

$$P(n, k) = \frac{n!}{(n-k)!}$$

**ç‰¹æ®Šæƒ…æ³**ï¼š
- **å…¨æ’åˆ—**ï¼šç•¶ $k = n$ æ™‚ï¼Œ$P(n, n) = n!$
- **éšä¹˜ (Factorial)**ï¼š$n! = n \times (n-1) \times (n-2) \times \cdots \times 2 \times 1$
- æ…£ä¾‹ï¼š$0! = 1$

### ğŸ² ä¾‹å­ 1ï¼šå¯†ç¢¼é–

æœ‰ 5 å€‹ä¸åŒçš„æ•¸å­— {1, 2, 3, 4, 5}ï¼Œè¦æ’æˆ 3 ä½æ•¸çš„å¯†ç¢¼ï¼ˆä¸å¯é‡è¤‡ï¼‰ï¼Œæœ‰å¹¾ç¨®æ’æ³•ï¼Ÿ

$$P(5, 3) = \frac{5!}{(5-3)!} = \frac{5!}{2!} = \frac{120}{2} = 60$$

### ğŸ² ä¾‹å­ 2ï¼šè³½è·‘æ’å

10 å€‹äººåƒåŠ è³½è·‘ï¼Œå‰ 3 åï¼ˆé‡‘ã€éŠ€ã€éŠ…ï¼‰æœ‰å¹¾ç¨®å¯èƒ½ï¼Ÿ

$$P(10, 3) = \frac{10!}{7!} = 10 \times 9 \times 8 = 720$$

```python
# ============================================
# 2. æ’åˆ— (Permutations)
# ============================================

def permutation(n: int, k: int) -> int:
    """
    è¨ˆç®—æ’åˆ—æ•¸ P(n, k) = n! / (n-k)!
    
    Args:
        n: ç¸½å…±æœ‰å¹¾å€‹ç‰©ä»¶
        k: è¦é¸å¹¾å€‹é€²è¡Œæ’åˆ—
    
    Returns:
        æ’åˆ—æ•¸é‡
    """
    return math.factorial(n) // math.factorial(n - k)

print("\n" + "=" * 60)
print("ğŸ”¢ æ’åˆ— (Permutations)")
print("=" * 60)

# ä¾‹å­ 1ï¼šå¯†ç¢¼é–
n1, k1 = 5, 3
result1 = permutation(n1, k1)
print(f"\nä¾‹å­ 1ï¼šå¯†ç¢¼é–")
print(f"å¾ {n1} å€‹æ•¸å­—ä¸­é¸ {k1} å€‹æ’åˆ—ï¼ˆä¸é‡è¤‡ï¼‰")
print(f"P({n1}, {k1}) = {n1}! / ({n1}-{k1})! = {result1}")

# ä½¿ç”¨ itertools é©—è­‰
digits = [1, 2, 3, 4, 5]
perms = list(itertools.permutations(digits, k1))
print(f"å¯¦éš›åˆ—èˆ‰ï¼š{len(perms)} ç¨®")
print(f"å‰ 5 ç¨®çµ„åˆï¼š{perms[:5]}")

# ä¾‹å­ 2ï¼šè³½è·‘æ’å
n2, k2 = 10, 3
result2 = permutation(n2, k2)
print(f"\nä¾‹å­ 2ï¼šè³½è·‘æ’å")
print(f"å¾ {n2} å€‹äººä¸­é¸å‡ºå‰ {k2} åï¼ˆé‡‘éŠ€éŠ…ï¼‰")
print(f"P({n2}, {k2}) = {n2}! / ({n2}-{k2})! = {result2}")
print(f"ä¹Ÿå¯ä»¥é€™æ¨£ç®—ï¼š{n2} Ã— {n2-1} Ã— {n2-2} = {n2 * (n2-1) * (n2-2)}")

# ä¾‹å­ 3ï¼šå…¨æ’åˆ—
n3 = 4
result3 = math.factorial(n3)
print(f"\nä¾‹å­ 3ï¼šå…¨æ’åˆ—")
print(f"4 å€‹å­—æ¯ {{A, B, C, D}} çš„æ‰€æœ‰æ’åˆ—æ–¹å¼")
print(f"P({n3}, {n3}) = {n3}! = {result3}")

# ä½¿ç”¨ itertools é©—è­‰ä¸¦åˆ—å‡º
letters = ['A', 'B', 'C', 'D']
all_perms = list(itertools.permutations(letters))
print(f"\næ‰€æœ‰ {len(all_perms)} ç¨®æ’åˆ—ï¼š")
for i, perm in enumerate(all_perms, 1):
    print(f"{i:2d}. {''.join(perm)}", end="  ")
    if i % 6 == 0:
        print()  # æ›è¡Œ
print("\n" + "=" * 60)
```

## 3. çµ„åˆ (Combinations)

### ğŸ“– ç†è«–

**çµ„åˆ (Combination)**ï¼šå¾ $n$ å€‹ä¸åŒçš„ç‰©ä»¶ä¸­ï¼Œé¸å‡º $k$ å€‹ï¼ˆé †åºä¸é‡è¦ï¼‰ã€‚

$$C(n, k) = \binom{n}{k} = \frac{n!}{k!(n-k)!}$$

**è¨˜è™Ÿ**ï¼š
- $C(n, k)$ æˆ– $\binom{n}{k}$ è®€ä½œ "n choose k"
- ä¹Ÿå¯«ä½œ $nCk$ æˆ– $_nC_k$

**é—œéµå·®ç•°**ï¼š
- **æ’åˆ— (Permutation)**ï¼šé †åºé‡è¦ â†’ {A, B, C} å’Œ {B, A, C} æ˜¯ä¸åŒçš„
- **çµ„åˆ (Combination)**ï¼šé †åºä¸é‡è¦ â†’ {A, B, C} å’Œ {B, A, C} æ˜¯åŒä¸€å€‹

**é—œä¿‚å¼**ï¼š
$$C(n, k) = \frac{P(n, k)}{k!}$$

å› ç‚ºæ’åˆ—æœƒæŠŠåŒæ¨£çš„ $k$ å€‹å…ƒç´ çš„ $k!$ ç¨®æ’åˆ—éƒ½ç®—é€²å»ï¼Œæ‰€ä»¥è¦é™¤ä»¥ $k!$ã€‚

### ğŸ² ä¾‹å­ 1ï¼šæŠ½ç

å¾ 10 å€‹äººä¸­æŠ½ 3 å€‹äººå¾—çï¼ˆçå“ç›¸åŒï¼‰ï¼Œæœ‰å¹¾ç¨®å¯èƒ½ï¼Ÿ

$$C(10, 3) = \frac{10!}{3! \times 7!} = \frac{10 \times 9 \times 8}{3 \times 2 \times 1} = \frac{720}{6} = 120$$

### ğŸ² ä¾‹å­ 2ï¼šæ’²å…‹ç‰Œ

å¾ 52 å¼µæ’²å…‹ç‰Œä¸­æŠ½ 5 å¼µï¼Œæœ‰å¹¾ç¨®å¯èƒ½ï¼Ÿ

$$C(52, 5) = \frac{52!}{5! \times 47!} = 2{,}598{,}960$$

### ğŸ² ä¾‹å­ 3ï¼šæ©Ÿå™¨å­¸ç¿’æ‡‰ç”¨

å‡è¨­ä½ æœ‰ 1000 å¼µåœ–ç‰‡ï¼Œè¦é¸ 100 å¼µä½œç‚ºæ¸¬è©¦é›†ï¼ˆé †åºä¸é‡è¦ï¼‰ï¼Œæœ‰å¹¾ç¨®é¸æ³•ï¼Ÿ

$$C(1000, 100) = \frac{1000!}{100! \times 900!} \approx 6.39 \times 10^{139}$$

ï¼ˆå¤©æ–‡æ•¸å­—ï¼é€™å°±æ˜¯ç‚ºä»€éº¼æˆ‘å€‘ç”¨ random seed ä¾†å›ºå®šè³‡æ–™åˆ†å‰²ï¼‰

```python
# ============================================
# 3. çµ„åˆ (Combinations)
# ============================================

def combination(n: int, k: int) -> int:
    """
    è¨ˆç®—çµ„åˆæ•¸ C(n, k) = n! / (k! * (n-k)!)
    
    Args:
        n: ç¸½å…±æœ‰å¹¾å€‹ç‰©ä»¶
        k: è¦é¸å¹¾å€‹ï¼ˆé †åºä¸é‡è¦ï¼‰
    
    Returns:
        çµ„åˆæ•¸é‡
    """
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))

print("\n" + "=" * 60)
print("ğŸ¯ çµ„åˆ (Combinations)")
print("=" * 60)

# ä¾‹å­ 1ï¼šæŠ½ç
n1, k1 = 10, 3
result1 = combination(n1, k1)
print(f"\nä¾‹å­ 1ï¼šæŠ½ç")
print(f"å¾ {n1} å€‹äººä¸­æŠ½ {k1} å€‹äººå¾—çï¼ˆçå“ç›¸åŒï¼‰")
print(f"C({n1}, {k1}) = {n1}! / ({k1}! Ã— {n1-k1}!) = {result1}")

# ä½¿ç”¨ itertools é©—è­‰
people = list(range(1, n1 + 1))
combs = list(itertools.combinations(people, k1))
print(f"å¯¦éš›åˆ—èˆ‰ï¼š{len(combs)} ç¨®")
print(f"å‰ 10 ç¨®çµ„åˆï¼š{combs[:10]}")

# ä¾‹å­ 2ï¼šæ’²å…‹ç‰Œ
n2, k2 = 52, 5
result2 = combination(n2, k2)
print(f"\nä¾‹å­ 2ï¼šæ’²å…‹ç‰Œ")
print(f"å¾ {n2} å¼µç‰Œä¸­æŠ½ {k2} å¼µ")
print(f"C({n2}, {k2}) = {result2:,}")

# ä¾‹å­ 3ï¼šæ¯”è¼ƒæ’åˆ—èˆ‡çµ„åˆ
print(f"\nä¾‹å­ 3ï¼šæ’åˆ— vs çµ„åˆçš„å·®ç•°")
print(f"å¾ 5 å€‹å­—æ¯ {{A, B, C, D, E}} ä¸­é¸ 3 å€‹ï¼š")
print(f"  â€¢ æ’åˆ— P(5, 3) = {permutation(5, 3)} ï¼ˆé †åºé‡è¦ï¼‰")
print(f"  â€¢ çµ„åˆ C(5, 3) = {combination(5, 3)} ï¼ˆé †åºä¸é‡è¦ï¼‰")
print(f"  â€¢ é—œä¿‚ï¼šP(5, 3) = C(5, 3) Ã— 3! = {combination(5, 3)} Ã— {math.factorial(3)} = {combination(5, 3) * math.factorial(3)}")

# å¯¦éš›åˆ—èˆ‰çœ‹çœ‹å·®ç•°
letters = ['A', 'B', 'C', 'D', 'E']
perms_3 = list(itertools.permutations(letters, 3))
combs_3 = list(itertools.combinations(letters, 3))

print(f"\nçµ„åˆçš„ {len(combs_3)} ç¨®é¸æ³•ï¼š")
for i, comb in enumerate(combs_3, 1):
    print(f"{i:2d}. {set(comb)}")

print(f"\næ¯å€‹çµ„åˆå°æ‡‰ 3! = 6 ç¨®æ’åˆ—ï¼š")
print(f"ä¾‹å¦‚ {{A, B, C}} å°æ‡‰çš„ 6 ç¨®æ’åˆ—ï¼š")
example_perms = [p for p in perms_3 if set(p) == {'A', 'B', 'C'}]
for i, perm in enumerate(example_perms, 1):
    print(f"  {i}. {''.join(perm)}")

print("=" * 60)
```

## 4. ç¸½çµèˆ‡æ¯”è¼ƒ

### ğŸ“Š å¿«é€Ÿå°ç…§è¡¨

| å•é¡Œé¡å‹ | å…¬å¼ | é †åº | é‡è¤‡ | ä¾‹å­ |
|---------|------|------|------|------|
| **ä¹˜æ³•åŸç†** | $n_1 \times n_2 \times \cdots$ | âœ“ | âœ“ | å¯†ç¢¼ï¼ˆå¯é‡è¤‡ï¼‰ |
| **æ’åˆ— (Permutation)** | $P(n,k) = \frac{n!}{(n-k)!}$ | âœ“ | âœ— | è³½è·‘åæ¬¡ |
| **çµ„åˆ (Combination)** | $C(n,k) = \frac{n!}{k!(n-k)!}$ | âœ— | âœ— | æŠ½çã€å§”å“¡æœƒ |

### ğŸ¤” å¦‚ä½•åˆ¤æ–·ï¼Ÿ

**ä¸‰å€‹é—œéµå•é¡Œ**ï¼š
1. **é †åºé‡è¦å—ï¼Ÿ**
   - é‡è¦ â†’ è€ƒæ…®æ’åˆ— (Permutation)
   - ä¸é‡è¦ â†’ è€ƒæ…®çµ„åˆ (Combination)

2. **å¯ä»¥é‡è¤‡å—ï¼Ÿ**
   - å¯ä»¥ â†’ ä¹˜æ³•åŸç†
   - ä¸å¯ä»¥ â†’ æ’åˆ—æˆ–çµ„åˆ

3. **æ˜¯åˆ†æ­¥é©Ÿé€²è¡Œå—ï¼Ÿ**
   - æ˜¯ â†’ ä¹˜æ³•åŸç†
   - å¦ â†’ æ’åˆ—æˆ–çµ„åˆ

### ğŸ’¡ æ©Ÿå™¨å­¸ç¿’ä¸­çš„æ‡‰ç”¨

1. **è¶…åƒæ•¸èª¿æ•´ (Hyperparameter Tuning)**
   - å‡è¨­æœ‰ 3 å€‹ learning rates, 2 å€‹ batch sizes, 4 å€‹ optimizers
   - ç¸½å…±è¦æ¸¬è©¦ï¼š$3 \times 2 \times 4 = 24$ ç¨®çµ„åˆï¼ˆä¹˜æ³•åŸç†ï¼‰

2. **K-Fold Cross Validation**
   - å°‡è³‡æ–™åˆ†æˆ $k$ ä»½ï¼Œé¸ 1 ä»½ä½œç‚ºé©—è­‰é›†
   - æœ‰ $C(k, 1) = k$ ç¨®é¸æ³•ï¼ˆçµ„åˆï¼‰

3. **Feature Selection**
   - å¾ $n$ å€‹ç‰¹å¾µä¸­é¸ $k$ å€‹
   - ç¸½å…± $C(n, k)$ ç¨®çµ„åˆ

4. **Ensemble Methods**
   - å¾ 10 å€‹æ¨¡å‹ä¸­é¸ 3 å€‹é€²è¡Œ voting
   - $C(10, 3) = 120$ ç¨®çµ„åˆ

```python
# ============================================
# 4. å¯¦éš›æ‡‰ç”¨ç·´ç¿’ (Machine Learning Examples)
# ============================================

print("\n" + "=" * 60)
print("ğŸ¤– æ©Ÿå™¨å­¸ç¿’ä¸­çš„è¨ˆæ•¸æ‡‰ç”¨")
print("=" * 60)

# æ‡‰ç”¨ 1ï¼šè¶…åƒæ•¸ç¶²æ ¼æœç´¢
print("\næ‡‰ç”¨ 1ï¼šè¶…åƒæ•¸èª¿æ•´ (Grid Search)")
learning_rates = [0.001, 0.01, 0.1]
batch_sizes = [32, 64]
optimizers = ['Adam', 'SGD', 'RMSprop', 'AdaGrad']

total_configs = len(learning_rates) * len(batch_sizes) * len(optimizers)
print(f"Learning Rates: {len(learning_rates)} ç¨®")
print(f"Batch Sizes: {len(batch_sizes)} ç¨®")
print(f"Optimizers: {len(optimizers)} ç¨®")
print(f"ç¸½å…±éœ€è¦è¨“ç·´ï¼š{total_configs} å€‹æ¨¡å‹")

# åˆ—èˆ‰æ‰€æœ‰çµ„åˆ
all_configs = list(itertools.product(learning_rates, batch_sizes, optimizers))
print(f"\nå‰ 5 ç¨®çµ„åˆï¼š")
for i, (lr, bs, opt) in enumerate(all_configs[:5], 1):
    print(f"  {i}. LR={lr}, BatchSize={bs}, Optimizer={opt}")

# æ‡‰ç”¨ 2ï¼šK-Fold Cross Validation
print(f"\næ‡‰ç”¨ 2ï¼šK-Fold Cross Validation")
k = 5
print(f"å°‡è³‡æ–™åˆ†æˆ {k} ä»½ï¼Œæ¯æ¬¡é¸ 1 ä»½ä½œç‚ºé©—è­‰é›†")
print(f"ç¸½å…±æœ‰ C({k}, 1) = {combination(k, 1)} ç¨®åˆ†æ³•")
print(f"ä½†å¯¦éš›ä¸Šæˆ‘å€‘æœƒåš {k} æ¬¡è¨“ç·´ï¼ˆæ¯ä»½éƒ½ç•¶ä¸€æ¬¡é©—è­‰é›†ï¼‰")

# æ‡‰ç”¨ 3ï¼šFeature Selection
print(f"\næ‡‰ç”¨ 3ï¼šç‰¹å¾µé¸æ“‡ (Feature Selection)")
n_features = 10
select_k = 5
n_combinations = combination(n_features, select_k)
print(f"å¾ {n_features} å€‹ç‰¹å¾µä¸­é¸ {select_k} å€‹")
print(f"ç¸½å…±æœ‰ C({n_features}, {select_k}) = {n_combinations:,} ç¨®çµ„åˆ")
print(f"å¦‚æœæ¯å€‹çµ„åˆè¨“ç·´éœ€è¦ 1 ç§’ï¼Œç¸½å…±éœ€è¦ {n_combinations / 60:.1f} åˆ†é˜")

# æ‡‰ç”¨ 4ï¼šEnsemble Voting
print(f"\næ‡‰ç”¨ 4ï¼šé›†æˆå­¸ç¿’ Voting")
n_models = 7
select_models = 3
n_ensembles = combination(n_models, select_models)
print(f"å¾ {n_models} å€‹åŸºç¤æ¨¡å‹ä¸­é¸ {select_models} å€‹é€²è¡ŒæŠ•ç¥¨")
print(f"ç¸½å…±æœ‰ C({n_models}, {select_models}) = {n_ensembles} ç¨®é›†æˆæ–¹å¼")

# åˆ—èˆ‰æ‰€æœ‰å¯èƒ½çš„é›†æˆçµ„åˆ
models = [f"Model_{i}" for i in range(1, n_models + 1)]
ensembles = list(itertools.combinations(models, select_models))
print(f"\næ‰€æœ‰ {len(ensembles)} ç¨®é›†æˆçµ„åˆï¼š")
for i, ensemble in enumerate(ensembles, 1):
    print(f"  {i:2d}. {{{', '.join(ensemble)}}}")

print("=" * 60)
```
